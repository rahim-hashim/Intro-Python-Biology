### 3. Methods

Methods are similar to functions, just that they come after the variable you want to use (rather than before). It is probably unnecessary to go into the details as to why, but throughout your use of Python, you will be using both, so best to get familiar with them.

As we used indices to get different elements of lists, we can also add things to the end of lists using `.append()`, as is used below.

````python
list4 = ['Why', 'are', 'methods', 'so']
append_word = 'annoying'

list4.append(append_word)
print(list4)
````

As you can see, in the first blank position `__.append(___)` is the list you want to add something to, and the second blank position is what you are adding. **Importantly, append only takes one element in the second blank, so you have to do it one by one (for now).** 

Additionally, with append, you didn't have to set a new variable once you appended something! **It automatically sets the old variable to be the newly-extended variable, which is different from functions.**

Other common methods that can be used are `.split(___)` (which splits strings by spaces into lists of words), and `.reverse(___)`, which I'm sure you can guess what it does. `split()` does not change the underlying variable, so you have to set a new one in order to maintain the changes.** 

Example below:

````python
test = 'In all the lands, he knew all the best halal trucks'
test_list = test.split()

print(test_list)
````

Not providing any value to `split()` auto-assigns it as an empty space, whereas if you provide a value, it splits based on that value.

````python
period_sentence = 'we.can.split.by.anything'
split_sentence = period_sentence.split('.')
print(split_sentence)
````

And as you can expect, `.join()` does the exact opposite of split. The syntax is important though, with `__.join(__)`, the first space is the character you want to join on, and the second is the string you are joining with that character.

````python
test_reverse_string = ' '.join(split_sentence)
print(test_reverse_string)
````

Split the following sentence by hyphens and join using underscores.

````python
hyphen_sentence = 'split-to-underscore'
underscore_sentence = '_'.join(hyphen_sentence.split('-'))
print(underscore_sentence)
````

### 4. Writing your own functions

In addition to using in-built Python functions and methods, [you can also generate your own functions](http://swcarpentry.github.io/python-novice-gapminder/16-writing-functions/index.html). The basic structure looks like this:

````python
def print_hello():
  print('Hello!')
````

You can see that `def` creates a function called `print_hello`, which takes in no arguments `()`.  In line 2 just calls the `print('Hello!)` statement. You can then "call" the function anywhere after.

````python
print_hello()
````

Another function we'll write below takes in an argument `myName`, and also returns, or sends back, a value from the function. Once you call return in a function, everything beyond it is ignored, given that you've already returned to the main script. You can see this here:

````python
def my_first_function(myName):
  sentence = 'My name is ' + myName
  return sentence
  print('Ignored')
````

````python
myName_sentence = my_first_function('Rahim')
````

We've now assigned a new variable, myName_sentence, which is the value returned from `my_first_function`.

````python
print(myName_sentence)
````

Write a function that, when given a string `root` and a non-negative int `n`, return a larger string that is `n` copies of the original `root`.

Example:
* string_times('Hi', 2) → 'HiHi'
* string_times('Hi', 3) → 'HiHiHi'
* string_times('Hi', 1) → 'Hi'

````python
def string_times(root, n):
  return_string = root*n
  return return_string

multi_root = string_times('Hi', 2)
print(multi_root)
````


Write a function that, when given two int values, returns double their sum as a float.

* sum_double(1, 2) → 6.0
* sum_double(3, 2) → 10.0
* sum_double(2, 2) → 8.0

````python
def sum_double(int1, int2):
  return_value = (int1 + int2) * 2.0
  return return_value

sum_double_return = sum_double(1,2)
print(sum_double_return)
````
